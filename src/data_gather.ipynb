```python
import numpy as np
import pandas as pd
import FinanceDataReader as fdr
```

```python
df_krx = fdr.StockListing('KRX-MARCAP')
df_krx
```

```python
import matplotlib.pyplot as plt

plt.plot(df_krx["Marcap"])
plt.show()
```

```python
num_stocks = 300

plt.plot(df_krx["Marcap"][:num_stocks])
plt.show()
```

```python
start_date = '2010-12-30'
end_date = '2021-12-30'
df_list = []

for code in df_krx["Code"][:num_stocks]:
    series_close = fdr.DataReader(code, start_date, end_date)["Close"]
    df_list.append(series_close)

df_close = pd.concat(df_list, axis=1)
df_close.columns = df_krx["Name"][:num_stocks]
df_close
```

```python
df_close = df_close.dropna(axis=1)
df_close
```

```python
df_returns = df_close.pct_change()
df_returns = df_returns.iloc[1:, :]
df_returns
```

```python
correlation_matrix = df_returns.corr()
correlation_matrix
```

```python
pd.options.display.float_format = '{:.5f}'.format
distance_matrix = 1-correlation_matrix
distance_matrix
```

```python
pip install sklearn
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(distance_matrix, 'ward')
linked
labelList = range(len(distance_matrix))

plt.figure(figsize=(20, 10))
dendrogram(
    linked,
    orientation='top',
    labels=labelList,
    distance_sort='descending',
    show_leaf_counts=True
)

plt.xticks(rotation=-90)
plt.tick_params(axis='x', labelcolor='white', bottom=True)
plt.tick_params(axis='y', labelcolor='white', top=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# TODO: how to choose num_clusters
num_clusters = 20

model = AgglomerativeClustering(n_clusters=num_clusters , affinity='euclidean', linkage='ward')
clst_labels = model.fit_predict(distance_matrix)
clst_labels
```

```python
df_clst_num = pd.DataFrame(clst_labels)
df_clst_num.columns = ["cluster_number"]
df_clst_num
```

```python
df_clst_num["cluster_number"].value_counts()
```

```python
df_for_silhouette = pd.DataFrame(data = {'cluster_number': clst_labels, 'silhouette_coefficient': silhouette})
df_for_silhouette.groupby('cluster_number').mean()
```
