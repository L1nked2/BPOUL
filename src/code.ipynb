# Clustering Method

## Hirarchical Clustering

### KRX data

```python
df_returns = pd.read_csv("../data/KRX_200_returns", index_col = 0)

correlation_matrix = df_returns.corr()
correlation_matrix
```

```python
pd.options.display.float_format = '{:.5f}'.format
distance_matrix = 1-correlation_matrix
distance_matrix
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(distance_matrix, 'ward')
linked
labelList = range(len(distance_matrix))

plt.figure(figsize=(20, 10))
dendrogram(
    linked,
    orientation='top',
    labels=labelList,
    distance_sort='descending',
    show_leaf_counts=True
)

plt.xticks(rotation=-90)
plt.tick_params(axis='x', labelcolor='white', bottom=True)
plt.tick_params(axis='y', labelcolor='white', top=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# TODO: how to choose num_clusters
num_clusters = 20

model = AgglomerativeClustering(n_clusters=num_clusters , affinity='euclidean', linkage='ward')
clst_labels = model.fit_predict(distance_matrix)
clst_labels
```

```python
df_clst_num = pd.DataFrame(clst_labels)
df_clst_num.columns = ["cluster_number"]
df_clst_num
```

```python
from sklearn.preprocessing import MinMaxScaler
# Algorithms
from minisom import MiniSom
from tslearn.barycenters import dtw_barycenter_averaging
from tslearn.clustering import TimeSeriesKMeans
from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
```

```python
normalized_df_close = (df_close-df_close.mean(axis=0))/df_close.std(axis=0)
normalized_df_close
```

```python
import random

rand_column_list = []
random.seed(30)
n_iter = 4
for iter in range(n_iter):
    rand_col_num = random.randint(0, normalized_df_close.shape[1])
    rand_column_list.append(rand_col_num)

rand_column_list
```

```python
# 한글 폰트 설정
plt.rc('font', family='NanumGothic') # For Windows
print(plt.rcParams['font.family'])
```

```python

plt.figure(figsize = (30, 20))

for i, col_num in enumerate(rand_column_list):
    
    
    normalized_close_by_time = normalized_df_close.iloc[:, col_num]    
    time = normalized_close_by_time.index
    
    plt.subplot(2, 2, i+1);
    plt.plot(time, normalized_close_by_time, 'r-')
    
    stock_name = normalized_df_close.columns[col_num]
    plt.title(f'{stock_name}', fontsize = 20)
    
    plt.xlabel('time', fontsize = 15)
    plt.ylabel('normalized returns(%)', fontsize = 15)
    plt.xticks(fontsize = 13)
    
    
    
plt.show()
```

```python
import math 

mySeries = []

for col_num in range(normalized_df_close.shape[1]):
    mySeries.append(normalized_df_close.iloc[col_num])
    
print(mySeries)

for i in range(len(mySeries)):
    scaler = MinMaxScaler()
    mySeries[i] = MinMaxScaler().fit_transform(mySeries[i])
    mySeries[i]= mySeries[i].reshape(len(mySeries[i]))


som_x = som_y = math.ceil(math.sqrt(math.sqrt(len(mySeries))))
# I didn't see its significance but to make the map square,
# I calculated square root of map size which is 
# the square root of the number of series
# for the row and column counts of som

som = MiniSom(som_x, som_y, len(mySeries[0]), sigma=0.3, learning_rate = 0.1)

som.random_weights_init(mySeries)
som.train(mySeries, n_iter)
```

```python
# Little handy function to plot series
def plot_som_series_averaged_center(som_x, som_y, win_map):
    fig, axs = plt.subplots(som_x,som_y,figsize=(100,100))
    fig.suptitle('Clusters')
    for x in range(som_x):
        for y in range(som_y):
            cluster = (x,y)
            if cluster in win_map.keys():
                for series in win_map[cluster]:
                    axs[cluster].plot(series,c="gray",alpha=0.5) 
                axs[cluster].plot(np.average(np.vstack(win_map[cluster]),axis=0),c="red")
            cluster_number = x*som_y+y+1
            axs[cluster].set_title(f"Cluster {cluster_number}")

    plt.show()
```

```python
win_map = som.win_map(mySeries)
# Returns the mapping of the winner nodes and inputs

plot_som_series_averaged_center(som_x, som_y, win_map)
```

## K-means

```python
cluster_count = math.ceil(math.sqrt(len(mySeries)))
print(cluster_count)
# A good rule of thumb is choosing k as the square root of the number of points in the training data set in kNN

km = TimeSeriesKMeans(n_clusters=cluster_count, metric="dtw")

labels = km.fit_predict(mySeries)
```

```python
plot_count = math.ceil(math.sqrt(cluster_count))

fig, axs = plt.subplots(plot_count,plot_count,figsize=(25,25))
fig.suptitle('Clusters')
row_i=0
column_j=0
# For each label there is,
# plots every series with that label
for label in set(labels):
    cluster = []
    for i in range(len(labels)):
            if(labels[i]==label):
                axs[row_i, column_j].plot(mySeries[i],c="gray",alpha=0.4)
                cluster.append(mySeries[i])
    if len(cluster) > 0:
        axs[row_i, column_j].plot(np.average(np.vstack(cluster),axis=0),c="red")
    axs[row_i, column_j].set_title("Cluster "+str(row_i*som_y+column_j))
    column_j+=1
    if column_j%plot_count == 0:
        row_i+=1
        column_j=0
        
plt.show()
```
